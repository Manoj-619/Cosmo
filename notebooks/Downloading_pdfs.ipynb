```python
def download_pdf(url, filename):
    """Downloads a PDF from the URL and saves it with the specified filename."""
    try:
        response = requests.get(url, timeout=10) 
        if response.status_code == 200:
            with open(filename, 'wb') as file:
                file.write(response.content)
            print(f"PDF downloaded successfully and saved as {filename}")
        else:
            print(f"Failed to download PDF from {url}. Status code: {response.status_code}")
    except Exception as e:
        print(f"An error occurred while downloading from {url}: {e}")


output_directory = r"C:\Users\smrit\Work\Kenpath\Ofqual"
os.makedirs(output_directory, exist_ok=True)

for index, row in df.iterrows():
    url = row['Link to PDF']
    
    # Only process valid PDF URLs
    if isinstance(url, str) and url.endswith('.pdf'):
        # Extract the filename from the URL
        filename_from_url = os.path.basename(url)
        filename = os.path.join(output_directory, filename_from_url)
        
        download_pdf(url, filename)
    else:
        print(f"Invalid or non-PDF URL: {url}")

print("Download process completed.")
```

```python
# Load the Excel file
input_file = r"C:\Users\smrit\Downloads\filtered_links.xlsx"
df = pd.read_excel(input_file)

# Ensure the output directory exists
output_directory = r"C:\Users\smrit\Work\Kenpath\Ofquals"
os.makedirs(output_directory, exist_ok=True)

# Add a 'Downloaded' column with default value False
df['Downloaded'] = False

# Function to download a PDF and update the DataFrame
def download_pdf(row):
    url = row['Link to PDF']
    if isinstance(url, str) and url.endswith('.pdf'):
        filename_from_url = os.path.basename(url)
        filename = os.path.join(output_directory, filename_from_url)
        try:
            response = requests.get(url, timeout=10)
            if response.status_code == 200:
                with open(filename, 'wb') as file:
                    file.write(response.content)
                df.loc[row.name, 'Downloaded'] = True  # Update to True if successful
            else:
                print(f"Failed to download PDF from {url}. Status code: {response.status_code}")
        except Exception as e:
            print(f"An error occurred while downloading from {url}: {e}")
    else:
        print(f"Invalid or non-PDF URL: {url}")

# Use ThreadPoolExecutor to download PDFs concurrently
with ThreadPoolExecutor(max_workers=10) as executor:
    list(tqdm(executor.map(download_pdf, [row for _, row in df.iterrows()]), total=len(df), desc="Downloading PDFs"))

# Save the updated DataFrame to a new Excel file
output_file = 'main_ofqual_with_downloaded_status.xlsx'
df.to_excel(output_file, index=False)

print(f"Download process completed. Updated file saved to {output_file}.")

```

### Parsing 

```python
pdf_folder = r"C:\Users\smrit\Downloads\ofqual_julliette"
output_excel = r"C:\Users\smrit\Downloads\NOS & OFQUAL\ofqual_3.xlsx"

data = []

pdf_files = [file_name for file_name in os.listdir(pdf_folder) if file_name.endswith(".pdf")]

failed_files = []

def clean_text(text):
    cleaned_text = ''.join(c for c in text if c in string.printable)
    return cleaned_text

for file_name in tqdm(pdf_files, desc="Processing PDFs"):
    pdf_path = os.path.join(pdf_folder, file_name)
    try:
        reader = PdfReader(pdf_path)

        # Extract the title from the PDF metadata if available
        metadata = reader.metadata
        title = metadata.get("/Title", "No Title Found") if metadata else "No Title Found"

        full_text = ""
        for page in reader.pages:
            page_text = page.extract_text()
            full_text += page_text

        cleaned_text = re.sub(r"\s+", " ", full_text.strip())
        cleaned_text = clean_text(cleaned_text)

        data.append({
            "UUID": str(uuid.uuid4()), 
            "Text": cleaned_text,
            "Filename": file_name,
            "Title": title  # Add the title to the data
        })
    
    except Exception as e:
        failed_files.append(file_name)
        print(f"Failed to process {file_name}: {str(e)}")

# Create a DataFrame
df = pd.DataFrame(data)

# Save to Excel
df.to_excel(output_excel, index=False)

print(f"Data saved to {output_excel}")
if failed_files:
    print(f"Failed to process the following files: {', '.join(failed_files)}")
```
