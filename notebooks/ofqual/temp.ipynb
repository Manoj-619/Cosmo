```python
import pandas as pd
from urllib.parse import urlparse

# Extract unique domain names
def get_domain(url):
    try:
        # Parse the URL and extract the netloc (domain)
        return urlparse(url).netloc
    except:
        return None

```

```python
domain_counts = new_df["specification"].apply(get_domain).value_counts()
print(domain_counts[25:80])
```
