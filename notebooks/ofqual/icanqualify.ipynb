```python
import os
import pandas as pd
import requests

# Replace '/' with '_'
icq_df["modified_id"] = icq_df["ofqual_id"].str.replace("/", "_")

# Base URL
base_url = "https://icanqualify.net/qualification.icq/"

# Folder to save PDFs
download_folder = r"C:\Users\amith\Kenpath\OFQUAL\icanqualify_pdfs"
os.makedirs(download_folder, exist_ok=True)

# Headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept": "application/pdf",  # Ensure we request a PDF
    "Referer": "https://icanqualify.net/",  # Simulate coming from the website
}

# Iterate over each modified ID and download PDF
for modified_id in icq_df[:10]["modified_id"]:
    pdf_url = f"{base_url}{modified_id}"
    pdf_name = os.path.join(download_folder, f"{modified_id}.pdf")

    try:
        response = requests.get(pdf_url, headers=headers, stream=True)
        
        # Check if response is a valid PDF
        content_type = response.headers.get("Content-Type", "")
        if response.status_code == 200 and "application/pdf" in content_type:
            with open(pdf_name, "wb") as pdf_file:
                for chunk in response.iter_content(chunk_size=1024):
                    pdf_file.write(chunk)
            print(f"Downloaded: {pdf_name}")
        else:
            print(f"Failed to download {pdf_url} (Invalid content: {content_type})")

    except Exception as e:
        print(f"Error downloading {pdf_url}: {e}")

print("Download process completed.")
```

```python
import pandas as pd
import requests
import re
from urllib.parse import urljoin
from bs4 import BeautifulSoup
import os
import time
from concurrent.futures import ThreadPoolExecutor, as_completed

def download_pdf_from_qualification_id(qualification_id, output_dir=r"C:\Users\amith\Kenpath\OFQUAL\icanqualify_pdfs"):
    # Replace slashes with underscores in the ID
    safe_id = qualification_id.replace('/', '_')
    
    # Construct the original URL
    original_url = f"https://icanqualify.net/qualification.icq/{qualification_id}"
    
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Default output filename
    output_filename = os.path.join(output_dir, f"{safe_id}-icanqualify.pdf")
    
    try:
        # Get the HTML content
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
        }
        response = requests.get(original_url, headers=headers)
        
        if response.status_code != 200:
            return f"Failed to access {original_url}: {response.status_code}"
        
        html_content = response.text
        
        # Extract base URL
        base_url = "https://icanqualify.net"
        
        # Find the PDF path in the HTML
        # Method 1: Using regex to find SRC attribute in FRAME tag
        pdf_path_match = re.search(r'<FRAME[^>]+SRC=[\'"]([^\'"]+\.pdf)[\'"]', html_content, re.IGNORECASE)
        
        if not pdf_path_match:
            # Method 2: Try using BeautifulSoup as a fallback
            soup = BeautifulSoup(html_content, 'html.parser')
            frame_tags = soup.find_all('frame')
            
            pdf_path = None
            for frame in frame_tags:
                src = frame.get('src', '')
                if src.lower().endswith('.pdf'):
                    pdf_path = src
                    break
                
            if not pdf_path:
                return f"ID {qualification_id}: Could not find PDF path in the HTML"
        else:
            pdf_path = pdf_path_match.group(1)
        
        # Construct full PDF URL
        pdf_url = urljoin(base_url, pdf_path)
        
        # Download the PDF
        pdf_response = requests.get(pdf_url, headers=headers, stream=True)
        
        # Verify it's a PDF
        content_type = pdf_response.headers.get("Content-Type", "")
        if pdf_response.status_code == 200 and "application/pdf" in content_type:
            with open(output_filename, "wb") as pdf_file:
                for chunk in pdf_response.iter_content(chunk_size=1024):
                    pdf_file.write(chunk)
            return f"ID {qualification_id}: Successfully downloaded to {output_filename}"
        else:
            return f"ID {qualification_id}: Failed to download PDF: {content_type}"
            
    except Exception as e:
        return f"ID {qualification_id}: Error: {str(e)}"

def process_qualification_ids_from_dataframe(df, id_column='ofqual_id', output_dir=r"C:\Users\amith\Kenpath\OFQUAL\icanqualify_pdfs", max_workers=5, batch_size=None):

    results = []
    
    # Determine how many rows to process
    if batch_size is not None:
        df_to_process = df.head(batch_size)
    else:
        df_to_process = df
    
    # Create a list of qualification IDs
    qualification_ids = [str(row[id_column]) for _, row in df_to_process.iterrows()]
    
    # Use ThreadPoolExecutor to download PDFs in parallel
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        # Submit all download tasks
        future_to_id = {
            executor.submit(download_pdf_from_qualification_id, qid, output_dir): qid
            for qid in qualification_ids
        }
        
        # Process results as they complete
        for future in as_completed(future_to_id):
            qid = future_to_id[future]
            try:
                result = future.result()
                results.append(result)
                print(result)
            except Exception as exc:
                results.append(f"ID {qid}: Generated an exception: {exc}")
                print(f"ID {qid}: Generated an exception: {exc}")
    
    return results

# Example usage
results = process_qualification_ids_from_dataframe(
    icq_df, 
    id_column='ofqual_id',
    max_workers=15,  # Adjust based on your system capabilities
    batch_size=None   # Process first 100 records, set to None for all records
)
```
