```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set up the driver (make sure you have the right WebDriver installed)
driver = webdriver.Chrome()  # Change to webdriver.Firefox() if using Firefox
driver.get("https://www.1st4sport.com/qualifications")

wait = WebDriverWait(driver, 10)

# Keep clicking "Load More" until it's no longer available
while True:
    try:
        load_more_button = wait.until(EC.element_to_be_clickable(
            (By.XPATH, '//*[@id="qualifications-search-results"]/div/div[2]/div/div[2]/button')
        ))
        load_more_button.click()
    except:
        break  # Button is no longer clickable

# Extract all qualification links
qualification_links = []
qualification_cards = driver.find_elements(By.CLASS_NAME, "QualificationResultsCard")

for card in qualification_cards:
    link_element = card.find_element(By.CLASS_NAME, "QualificationResultsCard__link")
    qualification_links.append(link_element.get_attribute("href"))

# Print all collected links
print(len(qualification_links))

# Close the browser
driver.quit()
```

```python
import os
import requests
from bs4 import BeautifulSoup

# Folder to save PDFs
download_folder = r"C:\Users\amith\Kenpath\OFQUAL\1st4sport_pdfs"
os.makedirs(download_folder, exist_ok=True)

# Function to download a PDF
def download_pdf(pdf_url):
    pdf_filename = os.path.join(download_folder, pdf_url.split("/")[-1] + ".pdf")
    response = requests.get(pdf_url)
    
    if response.status_code == 200:
        with open(pdf_filename, "wb") as pdf_file:
            pdf_file.write(response.content)
        print(f"Downloaded: {pdf_filename}")
    else:
        print(f"Failed to download: {pdf_url}")

# Visit each link and extract PDF URLs
for link in qualification_links:
    response = requests.get(link)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Find the PDF download link
        pdf_button = soup.select_one(".QualificationOverview__download .Button--download")
        
        if pdf_button:
            pdf_url = pdf_button["href"]
            if pdf_url.startswith("/"):  # Convert relative URL to absolute
                pdf_url = "https://www.1st4sport.com" + pdf_url
            
            # Download the PDF
            download_pdf(pdf_url)
        else:
            print(f"No PDF found on {link}")
    else:
        print(f"Failed to access {link}")

print("Process completed!")
```
